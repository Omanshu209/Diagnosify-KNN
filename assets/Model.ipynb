# Disease Prediction Model

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from math import sqrt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score
```

## Preparing the data

```python
df = pd.read_csv("training.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.drop('Unnamed: 133',axis = 1)
```

```python
# for training the final model
X = df.drop('prognosis',axis = 1).copy().values
y = df['prognosis'].copy().values

# for training the initial model to get the accuracy of the model
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

## Training the model

```python
n_neighbors = round(sqrt(df["prognosis"].unique().shape[0]))

if n_neighbors % 2 == 0:
    n_neighbors += 1
```

```python
model = KNeighborsClassifier(n_neighbors = n_neighbors)
```

### Initial model (trained using the training data)

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
#print(y_pred)
#print(y_test)
```

```python
print([model.score(X_test,y_test), 
       accuracy_score(y_test,y_pred), 
       precision_score(y_test,y_pred, labels = df["prognosis"].unique(), average = "micro"), 
       recall_score(y_test,y_pred, labels = df["prognosis"].unique(), average = "micro")]
)
```

### Final model (trained using the complete data)

```python
model.fit(X,y)
```

```python
model.score(X,y)
```
